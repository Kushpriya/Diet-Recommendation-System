```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
from sklearn.cluster import KMeans
import joblib
import csv

class VfcDietRecommendation:
    def __init__(self, age, height, weight, activity_level, body_type):
        self.age = age
        self.height = height
        self.weight = weight
        self.activity_level = activity_level
        self.body_type = body_type
        self.bmr = self.calculate_bmr()

    def calculate_bmr(self):
        if self.body_type == "endomorphic":
            return (10 * self.weight) + (6.25 * self.height) - (5 * self.age) - 161
        elif self.body_type == "ectomorphic":
            return (10 * self.weight) + (6.25 * self.height) - (5 * self.age) + 5
        elif self.body_type == "mesomorphic":
            return (10 * self.weight) + (6.25 * self.height) - (5 * self.age) + 100

    def calculate_calories(self, goal):
        bmr = self.bmr

        activity_levels = {
            "sedentary": 1.2,
            "lightly_active": 1.375,
            "moderately_active": 1.55,
            "very_active": 1.725,
            "extra_active": 1.9
        }

        activity_multiplier = activity_levels.get(self.activity_level, 1.0)

        bmr *= activity_multiplier

        if goal == "weight_loss":
            return bmr * 0.8
        elif goal == "weight_gain":
            return bmr * 1.2

    @staticmethod
    def load_food_items(csv_file):
        food_items = []
        with open(csv_file, 'r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                food_items.append(row)
        return food_items

    @staticmethod
    def select_meal(food_items, target_calories):
        closest_meal = min(food_items, key=lambda x: abs(int(x['Calories']) - target_calories))
        return closest_meal

    def recommend_diet(self, goal, breakfast_csv, lunch_csv, dinner_csv):
        total_calories = self.calculate_calories(goal)

        if goal == "weight_loss":
            breakfast_percentage = 0.3  # 30% of total calories
            lunch_percentage = 0.4  # 40% of total calories
            dinner_percentage = 0.3  # 30% of total calories
        elif goal == "weight_gain":
            breakfast_percentage = 0.25  # 25% of total calories
            lunch_percentage = 0.35  # 35% of total calories
            dinner_percentage = 0.4  # 40% of total calories

        breakfast_calories = total_calories * breakfast_percentage
        lunch_calories = total_calories * lunch_percentage
        dinner_calories = total_calories * dinner_percentage

        breakfast_items = self.load_food_items(breakfast_csv)
        lunch_items = self.load_food_items(lunch_csv)
        dinner_items = self.load_food_items(dinner_csv)

        recommended_diet = {
            "Breakfast": {"Items": [], "Total Calories": 0},
            "Lunch": {"Items": [], "Total Calories": 0},
            "Dinner": {"Items": [], "Total Calories": 0},
        }

        recommended_diet["Breakfast"]["Items"].append(self.select_meal(breakfast_items, breakfast_calories))
        recommended_diet["Lunch"]["Items"].append(self.select_meal(lunch_items, lunch_calories))
        recommended_diet["Dinner"]["Items"].append(self.select_meal(dinner_items, dinner_calories))

        for meal, items in recommended_diet.items():
            total_calories = sum(int(item["Calories"]) for item in items["Items"] if item)
            recommended_diet[meal]["Total Calories"] = total_calories

        return recommended_diet

class DietRecommendationSystem:
    def __init__(self, user):
        self.user = user
        self.breakfast_items = pd.read_csv("Data_Set/Breakfast_data.csv")
        self.lunch_items = pd.read_csv("Data_Set/Lunch_data.csv")
        self.dinner_items = pd.read_csv("Data_Set/Dinner_data.csv")

    @staticmethod
    def knn_recommendation(meal_items, target_calories, k=5):
        X = meal_items[['Calories']].values
        nbrs = NearestNeighbors(n_neighbors=k, algorithm='auto').fit(X)
        distances, indices = nbrs.kneighbors([[target_calories]])
        recommended_items = meal_items.iloc[indices[0]]
        return recommended_items

    @staticmethod
    def kmeans_recommendation(meal_items, target_calories, k=3):
        X = meal_items[['Calories']].values
        kmeans = KMeans(n_clusters=k, random_state=42).fit(X)
        cluster_centers = kmeans.cluster_centers_
        distances = [abs(calories - target_calories) for calories in cluster_centers[:, 0]]
        closest_cluster_index = distances.index(min(distances))
        cluster_items = meal_items[kmeans.labels_ == closest_cluster_index]
        return cluster_items

    def recommend_diet_knn(self, goal):
        total_calories = self.user.calculate_calories(goal)
        breakfast_calories = total_calories * 0.3
        lunch_calories = total_calories * 0.4
        dinner_calories = total_calories * 0.3

        recommended_diet = {
            "Breakfast": self.knn_recommendation(self.breakfast_items, breakfast_calories),
            "Lunch": self.knn_recommendation(self.lunch_items, lunch_calories),
            "Dinner": self.knn_recommendation(self.dinner_items, dinner_calories)
        }

        return recommended_diet

    def recommend_diet_kmeans(self, goal):
        total_calories = self.user.calculate_calories(goal)
        breakfast_calories = total_calories * 0.3
        lunch_calories = total_calories * 0.4
        dinner_calories = total_calories * 0.3

        recommended_diet = {
            "Breakfast": self.kmeans_recommendation(self.breakfast_items, breakfast_calories),
            "Lunch": self.kmeans_recommendation(self.lunch_items, lunch_calories),
            "Dinner": self.kmeans_recommendation(self.dinner_items, dinner_calories)
        }

        return recommended_diet

# Define the variables
age = 20
height = 167.64
weight = 54
activity_level = "moderately_active"
body_type = "ectomorphic"

# Usage
user = VfcDietRecommendation(age, height, weight, activity_level, body_type)
diet_system = DietRecommendationSystem(user)

# K-NN recommendation
recommended_diet_knn = diet_system.recommend_diet_knn("weight_gain")
print("\nRecommended Diet for Weight Gain (K-NN):")
for meal, items in recommended_diet_knn.items():
    print(meal + ":")
    print(items)

# K-Means recommendation
recommended_diet_kmeans = diet_system.recommend_diet_kmeans("weight_gain")
print("\nRecommended Diet for Weight Gain (K-Means):")
for meal, items in recommended_diet_kmeans.items():
    print(meal + ":")
    print(items)

# Save the K-NN model to a pickle file
knn_model = diet_system.knn_recommendation(diet_system.breakfast_items, 500)
joblib.dump(knn_model, 'knn_model.pkl')
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import os

# Load the dataset
df = pd.read_csv('Data_Set/Diet.csv')

# Separate into meal types
breakfast_data = df[df['Breakfast'] == 1].drop(columns=['Food_items', 'Breakfast'])
lunch_data = df[df['Lunch'] == 1].drop(columns=['Food_items', 'Lunch'])
dinner_data = df[df['Dinner'] == 1].drop(columns=['Food_items', 'Dinner'])

# Save processed data
if not os.path.exists('Data_Set'):
    os.makedirs('Data_Set')

breakfast_data.to_csv('Data_Set/Breakfast_data.csv', index=False)
lunch_data.to_csv('Data_Set/Lunch_data.csv', index=False)
dinner_data.to_csv('Data_Set/Dinner_data.csv', index=False)
```

```python
from sklearn.preprocessing import StandardScaler

def preprocess_data(meal_items):
    features = meal_items[['Calories', 'Fats', 'Proteins', 'Iron', 'Calcium', 'Sodium',
                           'Potassium', 'Carbohydrates', 'Fibre', 'VitaminD', 'Sugars']]
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(features)
    return scaled_features, scaler
```

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.cluster import KMeans
import joblib

class DietRecommendationSystem:
    def __init__(self):
        self.breakfast_items = pd.read_csv("Data_Set/Breakfast_data.csv")
        self.lunch_items = pd.read_csv("Data_Set/Lunch_data.csv")
        self.dinner_items = pd.read_csv("Data_Set/Dinner_data.csv")

    def train_knn_model(self, k=5):
        features, scaler = preprocess_data(self.breakfast_items)
        knn_model = NearestNeighbors(n_neighbors=k, algorithm='auto').fit(features)
        joblib.dump(knn_model, 'ML_model/knn_model.pkl')
        joblib.dump(scaler, 'ML_model/knn_scaler.pkl')  # Save the scaler for later use

    def train_kmeans_model(self, k=3):
        features, scaler = preprocess_data(self.breakfast_items)
        kmeans_model = KMeans(n_clusters=k, random_state=42).fit(features)
        joblib.dump(kmeans_model, 'ML_model/kmeans_model.pkl')
        joblib.dump(scaler, 'ML_model/kmeans_scaler.pkl')  # Save the scaler for later use

# Initialize the system
diet_system = DietRecommendationSystem()

# Create directory for models if it does not exist
if not os.path.exists('ML_model'):
    os.makedirs('ML_model')

# Train and save the models
diet_system.train_knn_model()
diet_system.train_kmeans_model()
```

```python
import joblib
import pandas as pd

# Load the models and scalers
knn_model = joblib.load('ML_model/knn_model.pkl')
kmeans_model = joblib.load('ML_model/kmeans_model.pkl')
knn_scaler = joblib.load('ML_model/knn_scaler.pkl')
kmeans_scaler = joblib.load('ML_model/kmeans_scaler.pkl')

# Load the test data (for example, breakfast items)
breakfast_items = pd.read_csv("Data_Set/Breakfast_data.csv")
```

```python
def preprocess_data(meal_items, scaler):
    features = meal_items[['Calories', 'Fats', 'Proteins', 'Iron', 'Calcium', 'Sodium',
                           'Potassium', 'Carbohydrates', 'Fibre', 'VitaminD', 'Sugars']]
    scaled_features = scaler.transform(features)
    return scaled_features

# Preprocess the test data
test_features_knn = preprocess_data(breakfast_items, knn_scaler)
test_features_kmeans = preprocess_data(breakfast_items, kmeans_scaler)
```
