```python
import pandas as pd
import numpy as np

class DietRecommendationSystem:
    def __init__(self, age, height, weight, activity_level, body_type, goal, veg_non_veg_pref, junk_food_pref):
        self.age = age
        self.height = height / 100  # Convert height from cm to meters
        self.weight = weight
        self.activity_level = activity_level
        self.body_type = body_type
        self.goal = goal
        self.veg_non_veg_pref = veg_non_veg_pref
        self.junk_food_pref = junk_food_pref
        self.bmr = self.calculate_bmr()
        self.caloric_needs = self.calculate_calories()
        self.adjusted_caloric_needs = self.adjust_caloric_needs()

    def calculate_bmr(self):
        if self.body_type == "endomorphic":
            return (10 * self.weight) + (6.25 * self.height * 100) - (5 * self.age) - 161
        elif self.body_type == "ectomorphic":
            return (10 * self.weight) + (6.25 * self.height * 100) - (5 * self.age) + 5
        elif self.body_type == "mesomorphic":
            return (10 * self.weight) + (6.25 * self.height * 100) - (5 * self.age) + 100

    def calculate_calories(self):
        bmr = self.bmr
        activity_multipliers = {
            "sedentary": 1.2,
            "lightly_active": 1.375,
            "moderately_active": 1.55,
            "very_active": 1.725,
            "extra_active": 1.9
        }
        return bmr * activity_multipliers.get(self.activity_level, 1.2)

    def adjust_caloric_needs(self):
        if self.goal == "weight_loss":
            return self.calculate_calories() - 500  # Typical deficit for weight loss
        elif self.goal == "weight_gain":
            return self.calculate_calories() + 500  # Typical surplus for weight gain
        else:  # For normal maintenance
            return self.calculate_calories()

    def get_recommendations(self, df):
        # Filter by user preferences
        filtered_df = df[(df['VegNovVeg'] == self.veg_non_veg_pref) & (df['Junk Food'] == self.junk_food_pref)]
        return filtered_df

    def select_best_items(self, df, target_calories):
        # Shuffle the DataFrame rows
        df_shuffled = df.sample(frac=1).reset_index(drop=True)
        selected_items = []
        total_calories = 0
        
        for _, row in df_shuffled.iterrows():
            if total_calories + row['Calories'] <= target_calories:
                selected_items.append(row)
                total_calories += row['Calories']
            if total_calories >= target_calories:
                break

        return selected_items, total_calories

    def recommend_meals(self, breakfast_df, lunch_df, dinner_df):
        proportions = {
            'normal': (0.30, 0.40, 0.30),
            'weight_loss': (0.30, 0.40, 0.30),
            'weight_gain': (0.25, 0.35, 0.40)
        }
        proportion = proportions[self.goal]
        breakfast_calories = self.adjusted_caloric_needs * proportion[0]
        lunch_calories = self.adjusted_caloric_needs * proportion[1]
        dinner_calories = self.adjusted_caloric_needs * proportion[2]

        daily_recommendations = []
        total_calories = 0

        # Select food items for each meal
        for meal_df, meal_type, target_calories in zip(
                [breakfast_df, lunch_df, dinner_df],
                ['Breakfast', 'Lunch', 'Dinner'],
                [breakfast_calories, lunch_calories, dinner_calories]):
            
            filtered_df = self.get_recommendations(meal_df)
            selected_items, meal_total_calories = self.select_best_items(filtered_df, target_calories)
            
            for item in selected_items:
                daily_recommendations.append({
                    'Meal': meal_type,
                    'Food Item': item['Food_items'],
                    'Calories': item['Calories'],
                    'Fats': item['Fats'],
                    'Proteins': item['Proteins'],
                    'Iron': item['Iron'],
                    'Calcium': item['Calcium'],
                    'Sodium': item['Sodium'],
                    'Potassium': item['Potassium'],
                    'Carbohydrates': item['Carbohydrates'],
                    'Fibre': item['Fibre'],
                    'VitaminD': item['VitaminD'],
                    'Sugars': item['Sugars']
                })
                
            total_calories += meal_total_calories

        return daily_recommendations, total_calories

# Load data
breakfast_df = pd.read_csv('Data_set/Breakfast.csv')
lunch_df = pd.read_csv('Data_set/Lunch.csv')
dinner_df = pd.read_csv('Data_set/Dinner.csv')

# User preferences
age = 25
height = 175
weight = 70
activity_level = 'moderately_active'
body_type = 'mesomorphic'
goal = 'normal'  # Options: 'weight_loss', 'weight_gain', 'normal'
veg_non_veg_pref = 0  # 0 for veg, 1 for non-veg
junk_food_pref = 0    # 0 for non-junk, 1 for junk

# Instantiate the recommendation system
diet_system = DietRecommendationSystem(age, height, weight, activity_level, body_type, goal, veg_non_veg_pref, junk_food_pref)

# Calculate daily calorie needs
calories_needed = diet_system.adjusted_caloric_needs
print(f"Calories needed per day: {calories_needed:.2f}")

# Get meal recommendations
daily_recommendations, total_calories = diet_system.recommend_meals(breakfast_df, lunch_df, dinner_df)

# Create DataFrame for recommendations
recommendations_df = pd.DataFrame(daily_recommendations)

# Add a total row using pd.concat
total_row = pd.DataFrame([{
    'Meal': 'Total',
    'Food Item': 'Total Calories',
    'Calories': total_calories
}])
recommendations_df = pd.concat([recommendations_df, total_row], ignore_index=True)

print("Meal Recommendations:")
print(recommendations_df)

# Optionally, save the recommendations to a CSV file
recommendations_df.to_csv('Data_set/Recommendations.csv', index=False)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Load data
breakfast_df = pd.read_csv('Data_set/Breakfast.csv')
lunch_df = pd.read_csv('Data_set/Lunch.csv')
dinner_df = pd.read_csv('Data_set/Dinner.csv')

# Combine data into one DataFrame
all_meals_df = pd.concat([breakfast_df, lunch_df, dinner_df], ignore_index=True)

# Add a 'Meal' column to distinguish between meals
all_meals_df['Meal'] = all_meals_df.apply(lambda x: 'Breakfast' if x.name < len(breakfast_df) else ('Lunch' if x.name < len(breakfast_df) + len(lunch_df) else 'Dinner'), axis=1)

# Define features and target variable
X = all_meals_df[['Calories', 'Fats', 'Proteins', 'Iron', 'Calcium', 'Sodium', 'Potassium', 'Carbohydrates', 'Fibre', 'VitaminD', 'Sugars']]
y = all_meals_df['Calories']  # Target variable for regression

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Initialize and train the Gradient Boosting Regressor
model = GradientBoostingRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Gradient Boosting Regressor:\n  Mean Squared Error: {mse:.2f}\n  R^2 Score: {r2:.2f}\n')
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
import joblib

# Load data
breakfast_df = pd.read_csv('Data_set/Breakfast.csv')
lunch_df = pd.read_csv('Data_set/Lunch.csv')
dinner_df = pd.read_csv('Data_set/Dinner.csv')

# Combine data into one DataFrame
all_meals_df = pd.concat([breakfast_df, lunch_df, dinner_df], ignore_index=True)

# Add a 'Meal' column to distinguish between meals
all_meals_df['Meal'] = all_meals_df.apply(lambda x: 'Breakfast' if x.name < len(breakfast_df) else ('Lunch' if x.name < len(breakfast_df) + len(lunch_df) else 'Dinner'), axis=1)

# Define features and target variable
X = all_meals_df[['Calories', 'Fats', 'Proteins', 'Iron', 'Calcium', 'Sodium', 'Potassium', 'Carbohydrates', 'Fibre', 'VitaminD', 'Sugars']]
y = all_meals_df['Calories']  # Target variable for regression

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Initialize and train the Gradient Boosting Regressor
model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"R^2 Score: {r2:.2f}")

# Save the model and scaler
joblib.dump(model, 'ML_Model/gradient_boosting_model.pkl')
joblib.dump(scaler, 'ML_Model/scaler.pkl')
```
